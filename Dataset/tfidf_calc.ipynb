```python
import nltk
import numpy as np
import random
import string 

import re
```

```python
corpus = []
for i in range(1,694):
    text_file = open("E:\AZ_Proj\scraping\Problemmod2\problemmod"+str(i)+".txt", "r")
 
    #read whole file to a string
    data = text_file.read()
    
    data = re.sub(r'\W',' ',data)
    data = re.sub(r'\s+',' ',data)
    corpus.append(data)
        
#     print(data)
    #close file
    text_file.close()
    
```

```python
print(corpus[0])
```

```python
# nltk.download('punkt')
wordfreq = {}
for i in range(1,694):
    text_file = open("E:\AZ_Proj\scraping\Problemmod2\problemmod"+str(i)+".txt", "r")
 
    #read whole file to a string
    data = text_file.read()
    
    tokens = nltk.word_tokenize(data)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1

    #close file
    text_file.close()
```

```python
print(wordfreq["ability"])
```

```python
wordfreqsorted = sorted(wordfreq)
```

```python
word_idf_values = {}
word_idf = {}
for document in corpus:
    for token in np.unique(nltk.word_tokenize(document)):
        if token not in word_idf_values.keys():
            word_idf_values[token] = 1
        else:
            word_idf_values[token] += 1


for token in word_idf_values.keys(): 
    word_idf[token]= (np.log(693/(1 +word_idf_values[token])))
```

```python
idfval = []
for token in wordfreqsorted:
    idfval.append(word_idf[token])
```

```python
word_to_idx = {}
counter = 0;
for token in wordfreqsorted:
    word_to_idx[token] = counter
    counter += 1
```

```python
word_to_idx
```

```python
from collections import Counter
doc =1
tf_idf = {}

for document in corpus:
    
    tokens = nltk.word_tokenize(document)
    
    counter = Counter(tokens)
    words_count = len(tokens)
    
    for token in np.unique(tokens):
        
        tf = counter[token]/words_count
        idf = word_idf[token]
        
        tf_idf[doc, word_to_idx[token]] = tf*idf

    doc += 1
```

```python
for key, value in tf_idf.items():
    print(key, ' : ', value)
```

```python
with open(r'E:\AZ_Proj\scraping\Problemmod2\tfidf.txt', 'w') as fp:
    for key, value in tf_idf.items():
        # write each item on a new line
        fp.write(str(key[0]) + ' ' + str(key[1]) + ' ' + str(value) +'\n') 
    print('Done')
```

```python
D = np.zeros((693, len(wordfreqsorted)))
for i in tf_idf:
    try:
        ind = i[1]
        D[i[0]-1][ind] = tf_idf[i]
    except:
        pass
```

```python
D
```

```python
magvec = []
for i in range(0,693):
    magvec.append(np.linalg.norm(D[i]))
```
